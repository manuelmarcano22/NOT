```python
from bokeh.plotting import output_notebook, figure, show
from bokeh.models import HoverTool, tools
from bokeh.models import Span, Label, Arrow, NormalHead
import numpy as np
output_notebook()
```

```python

# create a new plot
p = figure(plot_width=900, plot_height=500, title="Standard Star HD84937 (sdF) ",active_drag='pan', active_scroll='wheel_zoom')

#Tool to get wavelength
hover2 = HoverTool(
        tooltips=[
            ("(x,y)", "($x{1}, $y)"),
        ]
    )


#plot Lines


##Emission Lines
class line(object):
    def __init__(self,name):
        self.name = name
x = []
y = []
with open('standardnonfluxcal.txt') as file:
    for line in file:
        xval,yval =line.split()
        x.append(float(xval))
        y.append(float(yval))
    
# add some renderers

#change just some things about the x-axes
p.xaxis.axis_label = "Wavelength (A)"


p.line(x,y)





p.add_tools(hover2)
```

# Standard Star 

[HD 84937](http://www.ing.iac.es/Astronomy/observing/manuals/html_manuals/tech_notes/tn065-100/hd849.html). The two broad absroption bands should be due to O2 at 6562 A and  7605 A.
